## Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataframe["Age"]=le.fit_transform(dataframe.Age)
```

# Feature selection

```python
plt.figure(figsize=(40,40))
sns.heatmap(dataframe.corr(),annot=True,annot_kws={'size':15})
```

```python
def correlation(dataframe,threshold):
    col_corr=set()
    corr_matrix=dataframe.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr            
```

# Target and predictor seperation

```python
X=dataframe.iloc[: , :-1]
y=dataframe['class']
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
corr_feature=correlation(X_train,0.7)
len(set(corr_feature))
```

```python
corr_feature
```

```python
X_train.drop(corr_feature,axis=1)
X_test.drop(corr_feature,axis=1)
```

# Outlier detection

```python
plt.figure(figsize=(9,10),facecolor='white')
plotnumber=1
for column in dataframe.drop('class',axis=1):
    if plotnumber<35:
        ax=plt.subplot(5,7,plotnumber)
        sns.boxplot(dataframe[column])
        plt.xlabel(column,fontsize=9)
        plt.ylabel("count",fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

#### We dont perform outliers in this random forest because in random forest outlier doesnot effect our model 

# Model  for randomforest

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100, random_state=2)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report
y_pred=model.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

# Model for logistic regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=150)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
